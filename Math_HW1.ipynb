#### Задание:
Постройте на одном графике две кривые y(x) для функции двух переменной y(k, x) = cos(k * x), взяв для одной кривой значение k = 1, а для другой – любое другое k, не равное 1.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# y(k, x) = cos(k * x)
x = np.linspace(-np.pi, np.pi, 141)
k1 = 1
y1 = np.cos(k1 * x)
print('k1 =', k1)
k2 = int(input('Введите любое число, кроме 1: k2 = '))
y2 = np.cos(k2 * x)
plt.plot(x, y1, x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('tight')
plt.legend(('k1', 'k2'))
plt.grid(True)
plt.show()
```
