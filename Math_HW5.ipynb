```python
# Перестановку можно сделать при помощи модуля itertools.permutations, при n = k
```

```python
n = 3
for p in itertools.permutations("012", 3):
    print(''.join(str(x) for x in p))
print(f'Всего {math.factorial(n)} вариантов при k = 3')
```

```python
n = 8
for p in itertools.permutations("01234567", n):
    print(''.join(str(x) for x in p))
print(f'Всего {math.factorial(n)} вариантов при k = 8')
```

```python
n = 100
r = 0.7
x = np.random.rand(n)
y = r * x + (1 - r) * np.random.rand(n)

plt.plot(x, y, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

A = np.vstack([x, np.ones(len(x))]).T
a, b = np.linalg.lstsq(A, y)[0]

c = (np.sum((x) * (y))) / (np.sqrt((np.sum((x)**2)) * (np.sum((y)**2))))
print(f'Коэффициент корреляции x и y = {c}')

plt.plot([0, 1], [b, a + b])
plt.show()
```
